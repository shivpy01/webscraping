```python
import requests as req
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}

result = req.get("https://realpython.github.io/fake-jobs/", headers=headers)

print(result)
```

```python
bs = BeautifulSoup(result.text,'html.parser')
# bs.prettify()
```

```python
results = bs.find(id="ResultsContainer")
```

```python
# print(results.prettify())
```

```python
job_cards = results.find_all("div", class_="card-content")
# print(job_cards)
```

```python
for job_card in job_cards:
    title_element = job_card.find("h2", class_="title")
    company_element = job_card.find("h3", class_="company")
    location = job_card.find("p", class_="location")
    print(title_element.text)
    print(company_element.text)
    print(location.text.strip())
    print()

    f = open("jobs.txt", "a")
    f.write(f"{title_element.text.strip()}\n{company_element.text.strip()}\n{location.text}")
    f.close()
```
